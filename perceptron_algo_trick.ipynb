```python
# Import NumPy for numerical operations
import numpy as np

# Import Pandas for creating DataFrames
import pandas as pd

# Function to generate a classification dataset
from sklearn.datasets import make_classification

# Function to split data into train and test sets
from sklearn.model_selection import train_test_split

# Function to calculate accuracy
from sklearn.metrics import accuracy_score


# Generate a dataset
X, y = make_classification(

    # Total number of samples (rows)
    n_samples=100,

    # Total number of input features (columns)
    n_features=2,

    # Number of useful features
    n_informative=2,

    # Number of redundant features
    n_redundant=0,

    # Binary classification (0 and 1)
    n_classes=2,

    # Fix randomness so output remains same every run
    random_state=42
)


# Create DataFrame using feature matrix X
df = pd.DataFrame(

    # Data
    X,

    # Column names
    columns=['Feature1', 'Feature2']
)


# Add target column
df['Target'] = y


# Print first 5 rows
print(df.head())


# Split data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(

    # Features
    X,

    # Labels
    y,

    # 20% data for testing
    test_size=0.2,

    # Fix randomness
    random_state=42
)


# Initialize weight for Feature1 randomly
w1 = np.random.rand()

# Initialize weight for Feature2 randomly
w2 = np.random.rand()

# Initialize bias randomly
b = np.random.rand()


# Learning rate controls update size
lr = 0.1

# Number of passes over training data
epochs = 100


# Outer loop runs 100 times
for epoch in range(epochs):

    # Loop through every training sample
    for i in range(len(X_train)):

        # First feature value of current sample
        x1 = X_train[i][0]

        # Second feature value of current sample
        x2 = X_train[i][1]

        # Calculate weighted sum
        # z = w1*x1 + w2*x2 + bias
        z = w1 * x1 + w2 * x2 + b

        # Step Activation Function
        # If z >= 0 predict 1 else predict 0
        y_hat = 1 if z >= 0 else 0

        # Check if prediction is wrong
        if y_hat != y_train[i]:

            # Case 1:
            # Actual class is 1
            # Predicted class is 0
            if y_train[i] == 1:

                # Move decision boundary toward point

                # Update first weight
                w1 = w1 + lr * x1

                # Update second weight
                w2 = w2 + lr * x2

                # Increase bias
                b = b + lr

            # Case 2:
            # Actual class is 0
            # Predicted class is 1
            else:

                # Move decision boundary away from point

                # Decrease first weight
                w1 = w1 - lr * x1

                # Decrease second weight
                w2 = w2 - lr * x2

                # Decrease bias
                b = b - lr


# Empty list to store predictions
y_pred = []


# Loop through every test sample
for i in range(len(X_test)):

    # Compute weighted sum
    z = w1 * X_test[i][0] + w2 * X_test[i][1] + b

    # Apply step activation function
    prediction = 1 if z >= 0 else 0

    # Store prediction
    y_pred.append(prediction)


# Calculate accuracy
acc = accuracy_score(y_test, y_pred)


# Print final learned weights
print("Weight 1 =", w1)

# Print second weight
print("Weight 2 =", w2)

# Print learned bias
print("Bias =", b)

# Print model accuracy
print("Accuracy =", acc)
```
